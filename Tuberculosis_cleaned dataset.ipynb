```python
# Statistical Analysis
```

```python
# List of numerical columns
num_cols = ['Age', 'Fatigue', 'Weight_Loss']
```

```python
# Mean
mean_vals = df[num_cols].mean()
```

```python
# Median
median_vals = df[num_cols].median()
```

```python
# Mode (mode() returns a DataFrame, take first mode if multiple)
mode_vals = df[num_cols].mode().iloc[0]
```

```python
# Display all together
print("Mean Values:\n", mean_vals)
print("\nMedian Values:\n", median_vals)
print("\nMode Values:\n", mode_vals)
```

```python
col = 'Weight_Loss'

# Calculate standard deviation and variance
std_dev = df[col].std()
variance = df[col].var()

# Display results
print(f"Standard Deviation of {col}: {std_dev:.2f}")
print(f"Variance of {col}: {variance:.2f}")
```

```python
# If the mean weight loss is, say, 4.8 kg (just an example), and the standard deviation is 4.34 kg, it means:

# Most patients' weight loss falls in the range:
4.8 Â± 4.34 â†’ [0.46 kg to 9.14 kg]
```

```python
import matplotlib.pyplot as plt

# Plotting histogram for Fatigue
plt.figure(figsize=(8, 5))
plt.hist(df['Fatigue'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Fatigue Levels')
plt.xlabel('Fatigue Score')
plt.ylabel('Number of Patients')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Uniformly Distributed
# No skew
# Fatigue doesnâ€™t strongly differentiate TB from Normal
```

```python
# Calculating correlation between Weight_Loss and Fatigue
correlation = df['Weight_Loss'].corr(df['Fatigue'])

print(f"Correlation between Weight_Loss and Fatigue: {correlation:.2f}")
```

```python
# no linear correlation between Weight_Loss and Fatigue
```

```python
# Calculate correlation between Cough_Severity and Chest_Pain
correlation = df['Cough_Severity'].corr(df['Breathlessness'])

print(f"Correlation between Cough_Severity and Breathlessness: {correlation:.2f}")
```
